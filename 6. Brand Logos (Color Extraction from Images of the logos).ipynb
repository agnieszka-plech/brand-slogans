<h1>Brand Logos - Color Extraction from Images with Machine Learing application<span class="tocSkip"></span></h1>


In the last part of the project, I wanted to investigate brand logos, specifically the colors that are used by the companies from different categories. 




<h2>How does it work?<span class="tocSkip"></span></h2>



The images are first imported into a notebook as a 3D Numpy array (the first 2 values in the shape relate to the pixels of the image and the 3rd value to the representation of each pixel as a combination of 3 colors (RGB - Red, Green, Blue)).

In order to work with images, they need to be reshaped into a 2D array.

In the next step, RBG is converted into hex so that the color labels could be used in charts representing the colors.

After that, I built a K-means model with a selected number of clusters, which in this case are the top colors used in logos of brands from a specific category. 

With the use of Counter, I then got the count of all labels and transformed them back to the RGB and hex to use them in a plot.

## Converting RGB to Hex

First, we need to define a function that will convert RGB to hex so that we can use them as labels for our pie chart.

```python
def RGB2HEX(color):
    return "#{:02x}{:02x}{:02x}".format(int(color[0]), int(color[1]), int(color[2]))
```

## Model bulding: K-Means

K = TOTAL number of major colors that we want to extract from the image

```python
k = 6

# 1. Instantiate the model
kmeans = KMeans(n_clusters = k) 

# 2.Build the model with the image
kmeans.fit(final_images) 
```

```python
# Extract the labels (different cluster ids)
kmeans.labels_
```

```python
# we can also save it into a dataframe
clusters_ = pd.DataFrame(kmeans.labels_, columns=['cluster_id'])
```

```python
from glob import glob
import cv2
from numpy import asarray
from PIL import Image

# Load the folder with Airline logos:

final_images = np.zeros(shape= (1,3))
airlines_logos = glob('./Logos/Airlines/*.jpg', recursive=True)
for j in airlines_logos:
    img = Image.open(j)
    img = asarray(img)
    img = img.reshape((img.shape[0] * img.shape[1], img.shape[2]))
    final_images = np.concatenate((final_images,img), axis=0)
```

```python
final_images 
```

```python
final_images.shape
```

```python
img_init = final_images.copy() 
```

**Note:** 

I decided to extract 5 dominant colors, however, in each category white 
seemed to be have the highest percentage. The reason for that is the white background 
of each logo. I wanted to remove white, that's why the number of clusters is 6, not 5.


```python
k = 6

# 1. Instantiate the model
kmeans = KMeans(n_clusters = k) 

# 2.Build the model with the image
kmeans.fit(final_images) 
```

```python
kmeans.labels_
```

```python
hex_colors = [RGB2HEX(ordered_colors[i]) for i in counts]

hex_colors
```
